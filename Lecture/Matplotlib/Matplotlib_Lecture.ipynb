## 1. Matplotlib 란?

##### 개요

- 파이썬에서 자료를 차트(chart)나 플롯(plot)으로 시각화(visulaization)하는 패키지
- Pandas 시리즈와 데이터프레임의 plot 메소드 활용

#### 그릴 수 있는 Chart / Plot
- 라인 플롯(line plot)
- 산점도(scatter plot)
- 막대 그래프(bar chart)
- 히스토그램(histogram)
- 박스 플롯(box plot)
- 파이 차트(pie chart)
- 컨투어 플롯(contour plot)
- 서피스 플롯(surface plot)

- 틱 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

- grid 그리드 설정

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 표시")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(True)
plt.show()
```

- 여러개의 선을 그리기

- 범례

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label = "sine")
plt.legend(loc=0)
plt.show()
```

- x축 y축 라벨 타이틀

```python
X = np.linspace(-np.pi,np.pi,256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

여러가지 함수를 사용하여 아래 조건에 맞는 그래프를 그린다.

- xlabel, ylabel, title을 모두 갖추고 있어야 한다.
- 하나의 Figure(일단, 그림이라고 이해한다. 아래에 자세한 설명이 있다.)에 3개 이상의 Plot을 그린다.
- 각 Plot은 다른 선, 마크, 색 스타일을 가진다.
- legend는 그래프와 겹치지 않는 곳에 위치 시키도록 한다.


```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("복수의 plot 명령을 한그림에서 표현")
plt.plot(X, c = "b", lw=3, marker = "s", ms=3, label = "linspace")
plt.plot(C, c = "r", lw=2, marker = "o", ms=2, label = "cos(x)")
plt.plot(X**2, c = "m", lw=3, marker = "d", ms=3, label = "linspace^2")
plt.legend()
plt.xlabel("X 값")
plt.ylabel("Y 값")
```

- Axes 객체와 subplot 명령 

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)
plt.tight_layout()
plt.show()
```
