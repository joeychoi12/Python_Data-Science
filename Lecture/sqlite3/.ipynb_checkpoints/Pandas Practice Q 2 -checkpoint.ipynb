##### 4.6 데이터 프레임 합성 
###### 연습 문제 4.2.10
두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.

각각 5 x 5 이상의 크기를 가진다.
공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다.



```python
df1 = pd.DataFrame(
    np.random.randint(2,2000, size=(5,5)),
    index=['초','중','고','대학','대학원'],
    columns = ['서울','부산','대구','광주','대전'])
df1
```

```python
df2 = pd.DataFrame(
    np.random.randint(2000,5000, size=(5,5)),
    index=['유치원','초','중','고','대학'],
    columns = ['파리','베니스','베이징','광저우','고베'])
df2
```

```python
pd.merge(df1, df2, how='outer', left_index=True, right_index=True)
```

```python
df1.join(df2,how='outer')
```

##### 연습 문제 4.2.11
어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 실적 정보는 "매출", "비용", "이익" 으로 이루어진다. (이익 = 매출 - 비용).

또한 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
df1 = pd.DataFrame(
    np.random.randint(50,200, size=(6,2)),
    index=['1월','2월','3월','4월','5월','6월'],
    columns=['매출','비용'])
df1
df1['이익']=df1['매출']-df1['비용']
df1                        
```

```python
df2 = pd.DataFrame(
    np.random.randint(50,200, size=(6,2)),
    index=['7월','8월','9월','10월','11월','12월'],
    columns=['매출','비용'])
df2
df2['이익']=df2['매출']-df2['비용']
df2                        
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df3 = pd.concat([df3, pd.DataFrame({'매출': df3['매출'].sum(),
                                    '비용': df3['비용'].sum(),
                                    '이익': df3['이익'].sum()},
                                   index = ['총 실적'])], axis=0)
df3
```

##### 연습문제 1
key1의 값을 기준으로 data1의 값을 분류하여 합계를 구한 결과를 시리즈가 아닌 데이터프레임으로 구한다.



```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2

groups = df2.groupby('key1')
groups.sum()
```

##### 연습 문제 4
타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.



```python
titanic = sns.load_dataset("titanic")
```

1. qcut 명령으로 세 개의 나이 그룹을 만든다.
2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.
3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
#1. qcut 명령으로 세 개의 나이 그룹을 만든다.

bins = [1, 20, 60, 100]
labels = ["young", "middle age", "old"]
titanic_age = pd.cut(titanic["age"], bins, labels = labels)
titanic["age_class"] = titanic_age
titanic
```

```python
titanic.describe()
```

```python
#2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.
df1 = titanic.groupby(['sex', 'age_class', 'class'])["survived"].mean()
df2 = df1.unstack("class")
df2
```

```python
# 3. 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.
titanic.pivot_table(['survived'], index=['sex', 'class'], aggfunc='mean')
```
